```python
#usunięcie dwóch pierwszych kolumn
df = df.drop(columns = ["Unnamed: 0.1", "Unnamed: 0"])
```

```python
df.describe()
```

```python
#dane na histogramach względem wyników testu
for column in df.drop(columns = ["TEST_RESULT", "DATE_DIED", "SEX", "AGE_RANGE"]).columns:
    fig = plt.figure(figsize=(6,4))
    sns.histplot(data = df, x = column, bins=10, hue = "TEST_RESULT", multiple="dodge", palette="rocket_r")
    plt.grid()
    plt.show()
```

```python
#wykres gęstości - wyniki testu wg wieku
sns.set_context("paper",font_scale=1.1)
_, ax = plt.subplots(figsize=(12,6))
ax = sns.kdeplot(df.AGE[(df["TEST_RESULT"] == 0) ], color='orange', fill=True);
ax = sns.kdeplot(df.AGE[(df["TEST_RESULT"] == 1) ], color='purple', fill=True);
ax.legend(["Test result - Negative","Test result - Positive"],loc='upper right',fontsize=15);
ax.set_ylabel('Density');
ax.set_xlabel('AGE');
ax.set_title('Distribution of test results by age',fontdict={'fontsize':20,'fontweight':'bold'});
```

```python
#usunięcie rekordów z negatywnym wynikiem testu
df = df.drop(df[df.TEST_RESULT == 0].index)
```

```python
#usunięcie kolumny test result
df = df.drop(columns = ["TEST_RESULT"])
```

```python
#wykres korelacji
#widać sporą zależność między śmiercią a intubacją, zapaleniem płuc, wiekiem, cukrzycą, nadciśnieniem 
plt.figure(figsize=(8,6))
cor = df.corr(method="spearman") #korelacja spearmana - dla danych kategorycznych
sns.heatmap(cor, cmap=sns.color_palette("rocket_r", as_cmap=True))
plt.show()
```

```python
#dane na histogramach względem śmiertelności
for column in df.drop(columns = ["DEATH", "DATE_DIED", "SEX", "AGE_RANGE"] ).columns:
    fig = plt.figure(figsize=(6,4))
    sns.histplot(data = df, x = column, bins=10, hue = "DEATH", multiple="dodge", palette="rocket_r")
    plt.grid()
    plt.show()
```

```python
#wykres gęstości - śmiertelność wg wieku
sns.set_context("paper",font_scale=1.1)
_, ax = plt.subplots(figsize=(12,6))
ax = sns.kdeplot(df.AGE[(df["DEATH"] == 0) ], color='orange', fill=True);
ax = sns.kdeplot(df.AGE[(df["DEATH"] == 1) ], color='purple', fill=True);
ax.legend(["DEATH - No","DEATH - Yes"],loc='upper right',fontsize=15);
ax.set_ylabel('Density');
ax.set_xlabel('AGE');
ax.set_title('Distribution of deaths by age',fontdict={'fontsize':20,'fontweight':'bold'});
```

```python
#usuwam kolumny z najmniejszą korelacją
df2 = df.drop(columns = ['USMER', 'SEX', 'RETURNED_HOME', 'DATE_DIED', 'AGE', 'PREGNANT', 'COPD', 'ASTHMA', 'INMSUPR',  'OTHER_DISEASE','CARDIOVASCULAR', 'OBESITY', 'RENAL_CHRONIC', 'TOBACCO', 'ICU'])
df2.head()
```

```python
#zamiana rekordó na bardziej czytelne do aplikacji
df2['SEX_TYPE'] = df2['SEX_TYPE'].replace('F','Kobieta')
df2['SEX_TYPE'] = df2['SEX_TYPE'].replace('M','Mężczyzna')
df2['INTUBED'] = df2['INTUBED'].replace(0,'Nie')
df2['INTUBED'] = df2['INTUBED'].replace(1,'Tak')
df2['PNEUMONIA'] = df2['PNEUMONIA'].replace(0,'Nie')
df2['PNEUMONIA'] = df2['PNEUMONIA'].replace(1,'Tak')
df2['DIABETES'] = df2['DIABETES'].replace(0,'Nie')
df2['DIABETES'] = df2['DIABETES'].replace(1,'Tak')
df2['HIPERTENSION'] = df2['HIPERTENSION'].replace(0,'Nie')
df2['HIPERTENSION'] = df2['HIPERTENSION'].replace(1,'Tak')
df2['AGE_RANGE'] = df2['AGE_RANGE'].replace('AGE_0-18','0-18')
df2['AGE_RANGE'] = df2['AGE_RANGE'].replace('AGE_19-30','19-30')
df2['AGE_RANGE'] = df2['AGE_RANGE'].replace('AGE_31-40','31-40')
df2['AGE_RANGE'] = df2['AGE_RANGE'].replace('AGE_41-50','41-50')
df2['AGE_RANGE'] = df2['AGE_RANGE'].replace('AGE_51-60','51-60')
df2['AGE_RANGE'] = df2['AGE_RANGE'].replace('AGE_61-70','61-70')
df2['AGE_RANGE'] = df2['AGE_RANGE'].replace('AGE_71-80','71-80')
df2['AGE_RANGE'] = df2['AGE_RANGE'].replace('AGE_OVER_80','80+')
df2.head()
```

```python
#plik do aplikacji
df2.to_csv("covid_curr.csv")
```
