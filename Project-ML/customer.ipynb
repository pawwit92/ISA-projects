```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pd.set_option("display.max_columns", None)
df = pd.read_csv("testing_sample.csv")
df.head()
```

```python
df1 = pd.read_csv("training_sample.csv")
```

```python
df = df.drop(columns = ["UserID"])
df1 = df1.drop(columns = ["UserID"])
```

```python
df = pd.concat([df1, df])
```
